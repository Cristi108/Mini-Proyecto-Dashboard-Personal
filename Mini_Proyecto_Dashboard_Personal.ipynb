# Cargar librerías y dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/MCU Series.csv")
```

```python
df.info()
```

Gráfico 1 – IMDb Rating por Serie

```python
plt.figure(figsize=(10,5))
plt.plot(df["Series Title"], df["IMDb Ratings"], color='red', linestyle='--', marker='o')
plt.title("IMDb Rating por Serie")
plt.xticks(rotation=90, fontsize=6)
plt.ylabel("Rating")
plt.xlabel("Serie")
plt.tight_layout()
plt.savefig("imdb_por_serie.png")
plt.show()
```

Garfico 2 - Rotten Tomatoes por Serie

```python
plt.figure(figsize=(10,5))
plt.plot(df["Series Title"], df["Rotten Tomatoes Ratings (%)"], color='blue', linestyle='-', marker='x')
plt.title("Rotten Tomatoes por Serie")
plt.xticks(rotation=90, fontsize=6)
plt.ylabel("RT (%)")
plt.xlabel("Serie")
plt.tight_layout()
plt.savefig("rt_por_serie.png")
plt.show()
```

Grafico 3 - Número de Temporadas

```python
plt.figure(figsize=(10,5))
plt.plot(df["Series Title"], df["Seasons"], color='green', linestyle='-.', marker='s')
plt.title("Número de Temporadas por Serie")
plt.xticks(rotation=90, fontsize=6)
plt.ylabel("Temporadas")
plt.xlabel("Serie")
plt.tight_layout()
plt.savefig("temporadas_por_serie.png")
plt.show()
```

Grafico 4 - Número de Episodios

```python
plt.figure(figsize=(10,5))
plt.plot(df["Series Title"], df["Episodes"], color='purple', linestyle=':', marker='d')
plt.title("Número de Episodios por Serie")
plt.xticks(rotation=90, fontsize=6)
plt.ylabel("Episodios")
plt.xlabel("Serie")
plt.tight_layout()
plt.savefig("episodios_por_serie.png")
plt.show()
```

Grafico 5 - Series por Network (conteo)

```python
network_count = df["Network"].value_counts()

plt.figure(figsize=(10,5))
plt.plot(network_count.index, network_count.values, color='orange', linestyle='-', marker='^')
plt.title("Cantidad de Series por Network")
plt.xticks(rotation=90, fontsize=6)
plt.ylabel("Cantidad")
plt.xlabel("Network")
plt.tight_layout()
plt.savefig("series_por_network.png")
plt.show()
```

# Exploración del Dataset de Series

Después de analizar el dataset, pude ver varios patrones interesantes sobre las series y sus características principales. Aunque es un conjunto pequeño, ayuda a notar diferencias claras entre temporadas, episodios, ratings y los estudios o plataformas donde se transmitieron.

# Insights principales

1. **Las series con más temporadas no siempre tienen los mejores ratings.**
Algunas series largas tenían un IMDb más bajo que otras con menos temporadas.

2. **Rotten Tomatoes y IMDb no siempre coinciden.**
Hay series con un RT muy alto pero un IMDb más normal.

3. **Algunas networks aparecen más seguido.**
Se nota que ciertos canales o plataformas producen más series dentro del dataset.

4. **Hay series con pocos episodios pero muy buenas calificaciones.**
Esto muestra que no es necesario tener una serie larga para ser bien recibida.

5. **Las series más nuevas (en su estreno) tienden a tener mejores ratings.**
Aunque no es una regla, se observa un pequeño patrón.
